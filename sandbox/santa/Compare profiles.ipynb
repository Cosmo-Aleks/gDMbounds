# Comparison different profiles DM - $b\bar{b}$

### CTA - LMG

```python
# paper: Sensitivity of the Cherenkov Telescope Array to TeV photon emission from the Large Magellanic Cloud
# doi: 10.1093/mnras/stad1576
# instrument: CTA
# target: LMC
```

```python
from astropy.io import ascii
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d
```

```python
NFWmax_bb = ascii.read("../../dmbounds/bounds/cta/cta_2023_LMC_ann_bb_sens_NFWmax.ecsv")
NFWmean_bb = ascii.read("../../dmbounds/bounds/cta/cta_2023_LMC_ann_bb_sens_NFWmean.ecsv")
isomean_bb = ascii.read("../../dmbounds/bounds/cta/cta_2023_LMC_ann_bb_sens_ISOmean.ecsv")
isomax_bb = ascii.read("../../dmbounds/bounds/cta/cta_2023_LMC_ann_bb_sens_ISOmax.ecsv")
```

```python
plt.figure(figsize=(6,5))
plt.xlabel("DM mass [TeV]")
plt.ylabel(r"$\langle\sigma v \rangle$ cm$^{-3}$ s$^{-1}$")
plt.xlim(1e-1,2e2)
plt.ylim(1e-26,1e-23)
plt.title(r'Confronto di diversi profili di DM - $b\bar{b}$')
plt.loglog(NFWmax_bb['mass']*1e-3,NFWmax_bb['sigmav'],color='b',linestyle='dashed',label="NFWmax")
plt.loglog(NFWmean_bb['mass']*1e-3,NFWmean_bb['sigmav'],color='b',linestyle='dotted',label="NFWmean")
plt.loglog(isomax_bb['mass']*1e-3,isomax_bb['sigmav'],color='r',linestyle='dashed',label="ISOmax")
plt.loglog(isomean_bb['mass']*1e-3,isomean_bb['sigmav'],color='r',linestyle='dotted',label="ISOmean")


f = interp1d(isomean_bb['mass']*1e-3, isomean_bb['sigmav'])
y_iso_interp = f(isomax_bb['mass']*1e-3)
plt.fill_between(isomax_bb['mass']*1e-3,isomax_bb['sigmav'], y_iso_interp, color='red', alpha=0.1 )


g = interp1d(NFWmean_bb['mass']*1e-3, NFWmean_bb['sigmav'])
y_NFW_interp = g(NFWmax_bb['mass']*1e-3)
plt.fill_between(NFWmax_bb['mass']*1e-3,NFWmax_bb['sigmav'], y_NFW_interp, color='blue', alpha=0.1 )

# Thermal <sigma v>
x = np.linspace(1e-2, 2e2, 1000)
y_0 = 2.2e-26

plt.axhline(y=y_0, color='gray', linestyle='solid', label=r"Thermal $\langle \sigma v \rangle$")
plt.fill_between(x, y_0, color='lightgray', alpha=0.3)


plt.legend(loc="best",ncol=2)
plt.grid()

plt.savefig("./plot_thesis/profiles_cta_bb_LMC.eps")
# plt.savefig("./plot_thesis/profiles_cta_bb_LMC.eps")
# plt.savefig("./plot_thesis/profiles_cta_bb_LMC.eps")
```

```python
einasto_bb_hess = ascii.read("../../dmbounds/bounds/hess/hess_2015_gc_ann_qq_einasto.ecsv")
NFW_bb_hess = ascii.read("../../dmbounds/bounds/hess/hess_2015_gc_ann_qq_nfw.ecsv")
einasto_bb_cta = ascii.read("../../dmbounds/bounds/cta/cta_2015_gc_ann_bb_sens_einasto.ecsv")
NFW_bb_cta = ascii.read("../../dmbounds/bounds/cta/cta_2015_gc_ann_bb_sens_nfw.ecsv")
```

```python
plt.figure(figsize=(7,7))
plt.xlabel("DM mass [TeV]")
plt.ylabel(r"$\langle\sigma v \rangle$ cm$^{-3}$ s$^{-1}$")
plt.xlim(3e-2,15)
plt.ylim(1e-19,1e-27)
plt.title(r'Confronto di diversi profili DM - $b\bar{b}$')
plt.loglog(einasto_bb_hess['mass'],einasto_bb_hess['sigmav'],color='k',linestyle='dashed',label=r"Einasto $b\bar{b}$ HESS")
plt.loglog(NFW_bb_hess['mass'],NFW_bb_hess['sigmav'],color='b',linestyle='dotted',label=r"NFW $b\bar{b}$ HESS")
plt.loglog(einasto_bb_cta['mass']*1e-3,einasto_bb_cta['sigmav_lo'],color='y',linestyle='dashed',label=r"Einasto $b\bar{b}$ CTA")
plt.loglog(NFW_bb_cta['mass']*1e-3,NFW_bb_cta['sigmav_lo'],color='r',linestyle='dotted',label=r"NFW $b\bar{b}$ CTA")



plt.legend(loc="best",ncol=2)
plt.grid()

plt.savefig("profiles_hess_cta_bb_GC.eps")
# plt.savefig("profiles_hess_cta_bb_GC.pdf")
# plt.savefig("profiles_hess_cta_bb_GC.png")
```
