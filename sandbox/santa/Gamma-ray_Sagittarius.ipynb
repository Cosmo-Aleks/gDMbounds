# On the gamma-ray emission from the core of the Sagittarius dwarf galaxy

```python
from astropy.io import ascii
import matplotlib.pyplot as plt
import numpy as np
```

```python
gamma_Sagittarius = ascii.read("../../dmbounds/bounds/lat/lat_2023_sagittarius_ann_bb.ecsv")
```

```python
plt.figure(figsize=(7,7))
plt.xlabel("DM mass [TeV]")
plt.ylabel(r"$\langle\sigma v \rangle$ cm$^{-3}$ s$^{-1}$")
plt.xlim(1.3e-2,5e-2)
plt.ylim(1e-27,1e-25)
plt.title(r'$\gamma$-ray flux from Sagittarius dwarf galaxie')
plt.loglog(gamma_Sagittarius['mass']*1e-3,gamma_Sagittarius['sigmav'],color='k',linestyle='dashed',label=r"gamma-ray flux from Sgr dwarf galaxy$")


plt.legend(loc="best",ncol=2)
plt.grid()

#plt.savefig("Sagittarius.eps")
#plt.savefig("Sagittarius.pdf")
plt.savefig("Sagittarius.png")
```
