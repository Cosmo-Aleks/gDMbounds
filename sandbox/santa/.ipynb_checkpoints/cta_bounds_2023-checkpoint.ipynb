# Sensitivity of the Cherenkov Telescope Array to TeV photon emission from the Large Magellanic Cloud

```python
from astropy.io import ascii
import matplotlib.pyplot as plt
import numpy as np
```

## Comparison different channels - ISOmax model

```python
#ISOmax model
isomax_mumu = ascii.read("../../dmbounds/bounds/cta/cta_2023_LMC_ann_mumu_sens_ISOmax.ecsv")
isomax_bb = ascii.read("../../dmbounds/bounds/cta/cta_2023_LMC_ann_bb_sens_ISOmax.ecsv")
isomax_tautau = ascii.read("../../dmbounds/bounds/cta/cta_2023_LMC_ann_tautau_sens_ISOmax.ecsv")
isomax_WW = ascii.read("../../dmbounds/bounds/cta/cta_2023_LMC_ann_WW_sens_ISOmax.ecsv")
```

```python
plt.figure(figsize=(7,7))
plt.xlabel("DM mass [TeV]")
plt.ylabel(r"$\langle\sigma v \rangle$ cm$^{-3}$ s$^{-1}$")
plt.xlim(1e-1,2e2)
plt.ylim(1e-27,1e-23)
plt.title(r'Comparison different channels')
plt.loglog(isomax_mumu['mass']*1e-3,isomax_mumu['sigmav'],color='k',linestyle='dashed',label=r"ISOmax $\mu^{+}\mu^{-}$")
plt.loglog(isomax_bb['mass']*1e-3,isomax_bb['sigmav'],color='b',linestyle='dotted',label=r"ISOmax $b\bar{b}$")
plt.loglog(isomax_tautau['mass']*1e-3,isomax_tautau['sigmav'],color='r',linestyle='dashdot',label=r"ISOmax $\tau^{+}\tau^{-}$")
plt.loglog(isomax_WW['mass']*1e-3,isomax_WW['sigmav'],color='y',linestyle='dashdot',label=r"ISOmax $W^{+}W^{-}$")


plt.legend(loc="best",ncol=2)
plt.grid()

plt.savefig("compare_cta_ISOmax.eps")
plt.savefig("compare_cta_ISOmax.pdf")
plt.savefig("compare_cta_ISOmax.png")
```
