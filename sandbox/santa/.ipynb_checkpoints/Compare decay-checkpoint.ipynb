# Comparison DM decay

```python
from astropy.io import ascii
import matplotlib.pyplot as plt
import numpy as np
```

```python
perseus_cta_bb = ascii.read("../../dmbounds/bounds/cta/cta_2023_perseuscluster_dec_bb_sens.ecsv")
dsph_hawc_bb = ascii.read("../../dmbounds/bounds/hawc/hawc_2018_multidsph_dec_bb.ecsv")
dsph_lat_bb = ascii.read("../../dmbounds/bounds/lat/lat_2016_multidsph_dec_bb.ecsv")
dsph_lhaaso_bb = ascii.read("../../dmbounds/bounds/lhaaso/lhaaso_2020_multidsph_dec_bb_sens.ecsv")
perseus_magic_bb = ascii.read("../../dmbounds/bounds/magic/magic_2018_perseuscluster_dec_bb.ecsv")
segue1_magic_bb = ascii.read("../../dmbounds/bounds/magic/magic_2014_segue1_dec_bb.ecsv")
gc_magic_bb = ascii.read("../../dmbounds/bounds/magic/magic_2019_gchalo_dec_bb.ecsv")
```

```python
plt.figure(figsize=(8,6))
plt.xlabel("DM mass [TeV]", fontsize=12)
plt.ylabel(r"$\tau$ $[s]$", fontsize=12)
plt.xlim(8e-3,4e2)
plt.ylim(6e21,2e29)
plt.title(r'Decadimento in $b\bar{b}$', fontsize=16)
plt.loglog(perseus_cta_bb['mass'],perseus_cta_bb['tau'],color='k',linestyle='dashed',label=r"2023 CTA - Perseus")
plt.loglog(dsph_hawc_bb['mass'],dsph_hawc_bb['tau'],color='b',linestyle='solid',label=r"2018 HAWC - multidsph")
plt.loglog(dsph_lat_bb['mass']*1e-3,dsph_lat_bb['tau'],color='y',linestyle='solid',label=r"2016 FERMI-LAT - multidsph")
plt.loglog(dsph_lhaaso_bb['mass'],dsph_lhaaso_bb['tau'],color='m',linestyle='solid',label=r"2020 LHAASO - multidsph")
plt.loglog(perseus_magic_bb['mass'],perseus_magic_bb['tau'],color='r',linestyle='solid',label=r"2018 MAGIC - Perseus")
plt.loglog(segue1_magic_bb['mass'],segue1_magic_bb['tau'],color='g',linestyle='dashed',label=r"2014 MAGIC - Segue1")
plt.loglog(gc_magic_bb['mass']*1e-3,gc_magic_bb['tau'],color='c',linestyle='dashdot',label=r"2019 MAGIC - GC")


plt.legend(loc="best",ncol=2)
plt.grid()

plt.savefig("./plot_thesis/decay_bb.eps")
# plt.savefig("./plot_thesis/decay_bb.pdf")
# plt.savefig("./plot_thesis/decay_bb.png")
```

```python
perseus_cta_tautau = ascii.read("../../dmbounds/bounds/cta/cta_2023_perseuscluster_dec_tautau_sens.ecsv")
dsph_hawc_tautau = ascii.read("../../dmbounds/bounds/hawc/hawc_2018_multidsph_dec_tautau.ecsv")
dsph_lat_tautau = ascii.read("../../dmbounds/bounds/lat/lat_2016_multidsph_dec_tautau.ecsv")
dsph_lhaaso_tautau = ascii.read("../../dmbounds/bounds/lhaaso/lhaaso_2020_multidsph_dec_tautau_sens.ecsv")
perseus_magic_tautau = ascii.read("../../dmbounds/bounds/magic/magic_2018_perseuscluster_dec_tautau.ecsv")
segue1_magic_tautau = ascii.read("../../dmbounds/bounds/magic/magic_2014_segue1_dec_tautau.ecsv")
```

```python
plt.figure(figsize=(8,6))
plt.xlabel("DM mass [TeV]", fontsize=12)
plt.ylabel(r"$\tau$ $[s]$", fontsize=12)
plt.xlim(1e-3,1e3)
plt.ylim(1e23,2e29)
plt.title(r'Decadimento in $\tau^+ \tau^-$', fontsize=16)
plt.loglog(perseus_cta_tautau['mass'],perseus_cta_tautau['tau'],color='k',linestyle='dashed',label=r"2023 CTA - Perseus")
plt.loglog(dsph_hawc_tautau['mass'],dsph_hawc_tautau['tau'],color='b',linestyle='solid',label=r"2018 HAWC - multidsph")
plt.loglog(dsph_lat_tautau['mass']*1e-3,dsph_lat_tautau['tau'],color='y',linestyle='solid',label=r"2016 FERMI-LAT - multidsph")
plt.loglog(dsph_lhaaso_tautau['mass'],dsph_lhaaso_tautau['tau'],color='m',linestyle='solid',label=r"2020 LHAASO - multidsph")
plt.loglog(perseus_magic_tautau['mass']*1e-3,perseus_magic_tautau['tau'],color='r',linestyle='solid',label=r"2018 MAGIC - Perseus")
plt.loglog(segue1_magic_tautau['mass']*1e-3,segue1_magic_tautau['tau'],color='g',linestyle='dashed',label=r"2014 MAGIC - Segue1")



plt.legend(loc="best",ncol=2)
plt.grid()

plt.savefig("./plot_thesis/decay_tautau.eps")
# plt.savefig("./plot_thesis/decay_tautau.pdf")
# plt.savefig("./plot_thesis/decay_tautau.png")
```
