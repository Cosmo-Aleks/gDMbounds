# J factor

```python
from astropy.io import ascii
import matplotlib.pyplot as plt
import numpy as np
```

```python
sigue1_bb = ascii.read("../../dmbounds/bounds/cta/cta_2019_segue1_ann_bb_sens.ecsv")
gc_bb = ascii.read("../../dmbounds/bounds/cta/cta_2021_gc_ann_bb_sens.ecsv")
LMC_bb = ascii.read("../../dmbounds/bounds/cta/cta_2023_LMC_ann_bb_sens_NFWmean.ecsv")
perseus_bb = ascii.read("../../dmbounds/bounds/cta/cta_2023_perseuscluster_ann_bb_sens.ecsv")
subhalos_bb = ascii.read("../../dmbounds/bounds/cta/cta_2021_darksubhalos_ann_bb_sens_expo.ecsv")
```

```python
plt.figure(figsize=(6,5))
plt.xlabel("DM mass [TeV]", fontsize=12)
plt.ylabel(r"$\langle\sigma v \rangle$ [cm$^{3}$ s$^{-1}$]", fontsize=12)
plt.xlim(2e-2,2e2)
plt.ylim(5e-27,5e-21)
plt.title(r'CTA - $b\bar{b}$', fontsize=16)
plt.loglog(sigue1_bb['mass']*1e-3,sigue1_bb['sigmav'],color='k',linestyle='dashed',label="Sigue1")
plt.loglog(gc_bb['mass']*1e-3,gc_bb['sigmav'],color='b',linestyle='dotted',label="GC")
plt.loglog(LMC_bb['mass']*1e-3,LMC_bb['sigmav'],color='y',linestyle='dashdot',label="LMC")
plt.loglog(perseus_bb['mass']*1e-3,perseus_bb['sigmav'],color='m',linestyle='dotted',label="Perseus")
plt.loglog(subhalos_bb['mass']*1e-3,subhalos_bb['sigmav'],color='r',linestyle='dashdot',label="Dark subhaloes")

# Thermal <sigma v>
x = np.linspace(1e-2, 2e2, 1000)
y_0 = 3e-26

plt.axhline(y=y_0, color='gray', linestyle='solid', label=r"Thermal $\langle \sigma v \rangle$")
plt.fill_between(x, y_0, color='lightgray', alpha=0.3)


plt.legend(loc="best",ncol=2)
plt.grid()

plt.savefig("./plot_thesis/J_factors.eps")
# plt.savefig("./plot_thesis/J_factors.pdf")
# plt.savefig("./plot_thesis/J_factors.png")
```
