# Compare different experiments

```python
from astropy.io import ascii
import matplotlib.pyplot as plt
import numpy as np
```

```python
cta_bb = ascii.read("../../dmbounds/bounds/cta/cta_2023_LMC_ann_bb_sens_NFWmax.ecsv")
hess_bb = ascii.read("../../dmbounds/bounds/hess/hess_2022_gc_ann_bb.ecsv")
atlas_bb = ascii.read("../../dmbounds/bounds/collider/collider_2015_atlascms_ann_bb.ecsv")
directsearches_bb = ascii.read("../../dmbounds/bounds/directsearches/directsearches_2015_coupp_ann_bb.ecsv")
hawc_bb = ascii.read("../../dmbounds/bounds/hawc/hawc_2018_multidsph_ann_bb.ecsv")
lhaaso_bb = ascii.read("../../dmbounds/bounds/lhaaso/lhaaso_2019_multidsph_ann_bb_sens.ecsv")
lat_bb = ascii.read("../../dmbounds/bounds/lat/lat_2020_multidsph_ann_bb_oneD_frequentist.ecsv")
magic_bb = ascii.read("../../dmbounds/bounds/magic/magic_2022_segue1_ann_bb.ecsv")
magic_lat_bb = ascii.read("../../dmbounds/bounds/multi-inst/magic-lat_2016_segue1_ann_bb.ecsv")
veritas_bb = ascii.read("../../dmbounds/bounds/veritas/veritas_2017_multidsph_ann_bb.ecsv")
```

```python
plt.figure(figsize=(7,7))
plt.xlabel("DM mass [TeV]")
plt.ylabel(r"$\langle\sigma v \rangle$ cm$^{-3}$ s$^{-1}$")
plt.xlim(1e-4,2e2)
plt.ylim(1e-29,1e-21)
plt.title(r'Comparison different target - $b\bar{b}$')
plt.loglog(cta_bb['mass']*1e-3,cta_bb['sigmav'],color='black',linestyle=':',label=r"CTA")
plt.loglog(hess_bb['mass']*1e-3,hess_bb['sigmav'],color='blue',linestyle='--',label=r"HESS")
plt.loglog(atlas_bb['mass']*1e-3,atlas_bb['sigmav'],color='red',linestyle='-.',label=r"ATLAS CMS")
plt.loglog(directsearches_bb['mass']*1e-3,directsearches_bb['sigmav'],color='yellow',linestyle='-',label=r"Direct searches")
plt.loglog(hawc_bb['mass']*1e-3,hawc_bb['sigmav'],color='purple',linestyle=':',label=r"HAWC")
plt.loglog(lhaaso_bb['mass']*1e-3,lhaaso_bb['sigmav'],color='green',linestyle=':',label=r"HAWC")
plt.loglog(lat_bb['mass']*1e-3,lat_bb['sigmav'],color='orange',linestyle='-.',label=r"LAT")
plt.loglog(magic_bb['mass']*1e-3,magic_bb['sigmav'],color='pink',linestyle='-.',label=r"MAGIC")
plt.loglog(magic_lat_bb['mass']*1e-3,magic_lat_bb['sigmav'],color='magenta',linestyle='solid',label=r"LAT-MAGIC")
plt.loglog(veritas_bb['mass']*1e-3,veritas_bb['sigmav'],color='brown',linestyle=':',label=r"VERITAS")


plt.legend(loc="best",ncol=2)
plt.grid()

plt.savefig("./plot_thesis/compare_experiments_bb.eps")
#plt.savefig("./plot_thesis/compare_experiments_bb.pdf")
#plt.savefig("./plot_thesis/compare_experiments_bb.png")
```
