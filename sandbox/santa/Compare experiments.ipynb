# Compare different experiments

```python
from astropy.io import ascii
import matplotlib.pyplot as plt
import numpy as np
```

```python
cta_bb = ascii.read("../../dmbounds/bounds/cta/cta_2021_gc_ann_bb_sens.ecsv")
hess_bb = ascii.read("../../dmbounds/bounds/hess/hess_2022_gc_ann_bb.ecsv")
atlas_bb = ascii.read("../../dmbounds/bounds/collider/collider_2015_atlascms_ann_bb.ecsv")
directsearches_bb = ascii.read("../../dmbounds/bounds/directsearches/directsearches_2015_coupp_ann_bb.ecsv")
hawc_bb = ascii.read("../../dmbounds/bounds/hawc/hawc_2018_multidsph_ann_bb.ecsv")
lhaaso_bb = ascii.read("../../dmbounds/bounds/lhaaso/lhaaso_2019_multidsph_ann_bb_sens.ecsv")
lat_bb = ascii.read("../../dmbounds/bounds/lat/lat_2020_multidsph_ann_bb_oneD_frequentist.ecsv")
magic_bb = ascii.read("../../dmbounds/bounds/magic/magic_2022_multidsph_ann_bb.ecsv")
veritas_bb = ascii.read("../../dmbounds/bounds/veritas/veritas_2017_multidsph_ann_bb.ecsv")

#gamma_Sagittarius = ascii.read("../../dmbounds/bounds/lat/lat_2023_sagittarius_ann_bb.ecsv")
```

```python
plt.figure(figsize=(10,8))
plt.xlabel("DM mass [TeV]", fontsize=14)
plt.ylabel(r"$\langle\sigma v \rangle$ [cm$^{3}$ s$^{-1}$]", fontsize=14)
plt.xlim(1e-3,2e2)
plt.ylim(1e-29,1e-21)
plt.title(r'Confronto tra diversi esperimenti - $b\bar{b}$', fontsize=18)
plt.loglog(cta_bb['mass']*1e-3,cta_bb['sigmav'],color='black',linestyle='solid',label=r"2021 CTA - GC")
plt.loglog(hess_bb['mass'],hess_bb['sigmav'],color='blue',linestyle='dashdot',label=r"2022 HESS - GC")
plt.loglog(atlas_bb['mass']*1e-3,atlas_bb['sigmav'],color='red',linestyle='dashed',label=r"2015 ATLAS CMS")
plt.loglog(directsearches_bb['mass']*1e-3,directsearches_bb['sigmav'],color='yellow',linestyle='solid',label=r"2015 COUPP")
plt.loglog(hawc_bb['mass'],hawc_bb['sigmav'],color='purple',linestyle='dashdot',label=r"2018 HAWC - multidsph")
plt.loglog(lhaaso_bb['mass'],lhaaso_bb['sigmav'],color='green',linestyle='dashed',label=r"2019 LHAASO - multidsph")
plt.loglog(lat_bb['mass']*1e-3,lat_bb['sigmav'],color='orange',linestyle='solid',label=r"2020 FERMI-LAT - multidsph")
plt.loglog(magic_bb['mass']*1e-3,magic_bb['sigmav'],color='magenta',linestyle='solid',label=r"2022 MAGIC - multidsph")
plt.loglog(veritas_bb['mass']*1e-3,veritas_bb['sigmav'],color='brown',linestyle='dashdot',label=r"2017 VERITAS - multidsph")

#plt.loglog(gamma_Sagittarius['mass']*1e-3,gamma_Sagittarius['sigmav'],color='violet',linestyle='solid',label=r"2023 LAT - Sagittarius")


# Thermal <sigma v>
x = np.linspace(1e-4, 2e2, 1000)
y_0 = 3e-26

plt.axhline(y=y_0, color='gray', linestyle='solid', label=r"Thermal $\langle \sigma v \rangle$")
plt.fill_between(x, y_0, color='lightgray', alpha=0.3)

plt.legend(loc="best",ncol=2)
plt.grid()

plt.savefig("./plot_thesis/experiments_bb.eps")
#plt.savefig("./plot_thesis/experiments_bb.pdf")
#plt.savefig("./plot_thesis/experiments_bb.png")
```
