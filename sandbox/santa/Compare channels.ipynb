# Compare different channel 

### Comparison different channel - MAGIC-LAT, DWARF

```python
from astropy.io import ascii
import matplotlib.pyplot as plt
import numpy as np
```

```python
dsph_WW = ascii.read("../../dmbounds/bounds/multi-inst/magic-lat_2016_multidsph_ann_WW.ecsv")
dsph_bb = ascii.read("../../dmbounds/bounds/multi-inst/magic-lat_2016_multidsph_ann_bb.ecsv")
dsph_mumu = ascii.read("../../dmbounds/bounds/multi-inst/magic-lat_2016_multidsph_ann_mumu.ecsv")
dsph_tautau = ascii.read("../../dmbounds/bounds/multi-inst/magic-lat_2016_multidsph_ann_tautau.ecsv")
```

```python
plt.figure(figsize=(6,5))
plt.xlabel("DM mass [TeV]")
plt.ylabel(r"$\langle\sigma v \rangle$ cm$^{-3}$ s$^{-1}$")
plt.xlim(1e-2,1e2)
plt.ylim(1e-27,1e-21)
plt.title('Confronto di diversi canali di annichilazione - Dwarf galaxies')
plt.loglog(dsph_WW['mass']*1e-3,dsph_WW['sigmav'],color='k',linestyle='dashed',label=r"$W^+ W^-$")
plt.loglog(dsph_bb['mass']*1e-3,dsph_bb['sigmav'],color='b',linestyle='dotted',label=r"$b\bar{b}$")
plt.loglog(dsph_mumu['mass']*1e-3,dsph_mumu['sigmav'],color='r',linestyle='dashdot',label=r"$\mu^+ \mu^-$")
plt.loglog(dsph_tautau['mass']*1e-3,dsph_tautau['sigmav'],color='y',linestyle='dashdot',label=r"$\tau^+ \tau^-$")

# Thermal <sigma v>
x = np.linspace(1e-2, 2e2, 1000)
y_0 = 2.2e-26

plt.axhline(y=y_0, color='gray', linestyle='solid', label=r"Thermal $\langle \sigma v \rangle$")
plt.fill_between(x, y_0, color='lightgray', alpha=0.3)

plt.legend(loc="best",ncol=2)
plt.grid()

plt.savefig("./plot_thesis/channels_dsph.eps")
# plt.savefig("./plot_thesis/channels_dsph.pdf")
# plt.savefig("./plot_thesis/channels_dsph.png")
```

### Comparison different channels - CTA, LMC

```python
NFWmax_mumu = ascii.read("../../dmbounds/bounds/cta/cta_2023_LMC_ann_mumu_sens_NFWmax.ecsv")
NFWmax_bb = ascii.read("../../dmbounds/bounds/cta/cta_2023_LMC_ann_bb_sens_NFWmax.ecsv")
NFWmax_tautau = ascii.read("../../dmbounds/bounds/cta/cta_2023_LMC_ann_tautau_sens_NFWmax.ecsv")
NFWmax_WW = ascii.read("../../dmbounds/bounds/cta/cta_2023_LMC_ann_WW_sens_NFWmax.ecsv")
```
